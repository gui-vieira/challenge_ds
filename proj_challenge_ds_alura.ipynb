```python
# importação das bibliotecas utilizadas na análise
import pandas as pd
```

```python
# configurações de visualizações dos dataframes
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

```python
# leitura do arquivo JSON no git
df = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
df
```

```python
# ajuste das colunas que são dicionários, concatenação dos dados no df e exclusão das colunas duplicadas
phones = pd.json_normalize(df.phone)
customers = pd.json_normalize(df.customer)
internet_plan = pd.json_normalize(df.internet)
account_info = pd.json_normalize(df.account)

df = pd.concat([df, phones, customers, internet_plan, account_info], axis = 1)
df = df.drop(columns = ['customer','phone','internet','account'])
df
```
